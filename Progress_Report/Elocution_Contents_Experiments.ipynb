```python
# American Star Speaker
with open("ElocutionBooks/coo1-ark--13960-t58d0gk82-1682482537.txt", "r", encoding="utf-8") as text_file:
    text_1 = text_file.read()
  
text_1
```

```python
# Subset Text String to Table of Contents Area 

import re

def extract_contents(text):
    contents_pattern = r'\bCONTENTS\b'

    sections = re.split(contents_pattern, text, flags=re.IGNORECASE)

    if len(sections) > 1:
        # Extract the part after the "CONTENTS" heading
        table_of_contents = sections[1:(len(sections)-1)]
        return table_of_contents
    else:
        return None
```

```python
contents_text_1 = str(extract_contents(text_1))
contents_text_1
```

```python
type(contents_text_1)
```

```python
content_list = contents_text_1.split()
```

```python
content_nonull_list = []
for item in content_list:
    if item != "\\n":
        content_nonull_list.append(item)
```

```python
no_null_str = str(content_nonull_list)
```

```python
no_punct_srt = re.sub(r'[^\w\s]', '', no_null_str)
```

```python
no_punct_srt
```

```python
import nltk
from nltk.util import bigrams

tokens = nltk.word_tokenize(no_punct_srt)

bigrams_list = list(bigrams(tokens))
```

```python
bigrams_list
```
