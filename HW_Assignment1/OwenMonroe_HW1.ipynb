3. Perform Basic Transformations

```python
en_df['Text']
en_df['Lower Case Tweets'] = en_df['Text'].str.lower()

en_df[['Text', 'Lower Case Tweets']]
#https://sparkbyexamples.com/pandas/convert-pandas-column-to-lowercase/#:~:text=We%20can%20use%20str.,function%20on%20a%20specific%20column.
```

```python
def clean_hashtags(x):
    return re.findall(r'#(\S+)', str(x))

def clean_tweets_hash(y):
    return re.sub(r'#(\S+)', ' ', str(y))

def clean_tweets_acct(z):
    return re.sub(r'@(\S+)', ' ', str(z))

def clean_accounts(z):
    return re.findall(r'@(\S+)', str(z))
```

```python
en_df['Hashtags'] = en_df['Lower Case Tweets'].apply(clean_hashtags)
en_df['Accounts'] = en_df['Lower Case Tweets'].apply(clean_accounts)
en_df['CleanedTweets'] = en_df['Lower Case Tweets'].apply(clean_tweets_hash).apply(clean_tweets_acct)

en_df
```

Tokenizing Text with NLTK

```python
en_df['NLTK TokenizedTweets'] = en_df['CleanedTweets'].apply(nltk.word_tokenize)
en_df
```

Tokenizing Tweets with spaCy

```python
nlp = spacy.load("en_core_web_sm")
en_df['spacy TokenizedTweets'] = en_df['CleanedTweets'].apply(lambda x: nlp(x))
en_df
```

Removing Stopwords from Tokenized Tweets with NLTK

```python
stop_words = set(stopwords.words('english'))

en_df['StopwordRemovedTweets'] = en_df['NLTK TokenizedTweets'].apply(lambda x: ' '.join([word for word in x if word not in stop_words]))
en_df

# https://saturncloud.io/blog/how-to-remove-stop-words-from-a-pandas-dataframe-using-python/
```

Stemming the Removed Stopwords Tweets with NLTK

```python
stemmer = SnowballStemmer("english")
ps = PorterStemmer() 


def stemmer(word):
    str_word = str(word)
    return ps.stem(str_word)
```

```python
stemmer = SnowballStemmer('english')
en_df['NLTK Snowball StemmedTweets'] = en_df['StopwordRemovedTweets'].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split()]))
en_df
```
